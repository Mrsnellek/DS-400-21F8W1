```python
#Create a function called "years_working" that takes two agruments, "total_earnings" and "hourly_wages" 
# the function returns how many years it will take to earn the "total_earnings" working 40 hours a week, 
# 52 weeks a year at the given "hourly_wage"
def years_working(total_earnings, hourly_wage):
    return(total_earnings/(hourly_wage * 40 * 52))
```

```python
#How many years will it take to earn $100,000 working $7.25 per hour?
years_working(100000, 7.25)
```

```python
#How many years will it take to earn 129 billion dollars (Jeff Bezos' estimated wealth) 
# earning $100,000 per hour?
years_working(199*10**9, 100000)
```

## Chapter 9 Randomness

```python
# Create a function that simulates 10000 flips of a coin. 
# Record the number of heads and tails that are flipped in the simulation.
# Save the results in a data table.
hs= make_array('heads', 'tails')
def hs_outcome(games):
    heads_score = 0
    tails_score = 0
    for i in np.arange(games):
        flip = np.random.choice(hs)
        if flip == "heads":
            heads_score += 1
        else: tails_score += 1
    return Table().with_columns("Heads Score", heads_score, "Tails Score", tails_score)
hs_outcome(10000)
```

## Chapter 10: Sampling and Empirical Distributions

```python
# Create a plinko game where a "ball" is dropped from a height and it can either move right one space
# or left one space. The game should be wrapped in a function called "plinko_game". "plinko_game" should take
# 2 arguments: games (number of games to simulate) and height (height of the plinko game).  The function should
# return a table that records the game and the final distance the plinko ball taveled.
def plinko_game(games, height):
    end_spot = []
    moves = make_array(-1, 1)
    for i in np.arange(games):
        direction = []
        for i in np.arange(height):
            direction.append(np.random.choice(moves))
        end_spot.append(sum(direction))
    plinko_table = Table().with_columns("Game" , np.arange(games), "Distance", end_spot)
    return plinko_table
```

```python
plinko_table = plinko_game(10000, 15)
plinko_table
```

```python
# Plot the distribution of the distance the ball falls.
plinko_table.hist('Distance', bins = np.arange(-15,16,2))
```
